```python
# Cria uma amostra de coleÃ§Ã£o
users = {'Hans': 'active', 'Ã‰lÃ©onore': 'inactive', 'æ™¯å¤ªéƒ': 'active'}

# EstratÃ©gia: criar uma nova coleÃ§Ã£o
active_users = {}
for user, status in users.items():
    if status == 'active':
        active_users[user] = status
        
print(users, end="\n\n")
print(active_users)
```
